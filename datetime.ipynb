# time 

```python
# time import

import time
```

```python
# 현재시간 구하기

time.localtime()
```

```python
# 조금 더 깔끔하게 출력

print('Ymd: ',time.strftime('%Y-%m-%d', time.localtime())) # 년월일 출력
print('c: ',time.strftime('%c', time.localtime())) # 년월일시간 출력
print('a: ',time.strftime('%a', time.localtime())) # 요일 줄임
print('A: ',time.strftime('%A', time.localtime())) # 요일
print('w: ',time.strftime('%w', time.localtime())) # 요일 숫자 표시 0:일 ~6:토
print('d: ',time.strftime('%d', time.localtime())) # 일
print('b: ',time.strftime('%b', time.localtime())) # 월 줄임
print('B: ',time.strftime('%B', time.localtime())) # 월
print('m: ',time.strftime('%m', time.localtime())) # 월 숫자 표시
print('y: ',time.strftime('%y', time.localtime())) # 년도 두자리수
print('Y: ',time.strftime('%Y', time.localtime())) # 년도 네자리수
print('H: ',time.strftime('%H', time.localtime())) # 24시간
print('I: ',time.strftime('%I', time.localtime())) # 12시간
print('p: ',time.strftime('%p', time.localtime())) # AM, PM
print('M: ',time.strftime('%M', time.localtime())) # 분
print('S: ',time.strftime('%S', time.localtime())) # 초
print('z: ',time.strftime('%z', time.localtime())) # 대한민국 표준 시간대
print('j: ',time.strftime('%j', time.localtime())) # 1월1일로부터 경과된 일수
print('U: ',time.strftime('%U', time.localtime())) # 1년중 현재가 몇 주차인지
print('W: ',time.strftime('%W', time.localtime())) # 1년중 현재가 몇 주차인지
print('x: ',time.strftime('%x', time.localtime())) # 날짜 출력
print('X: ',time.strftime('%X', time.localtime())) # 시간 출력
```

# datetime

```python
# datetime import

import datetime as dt
```

```python
# 오늘 날짜 확인

dt.datetime.today()
```

```python
# 문자열을 시간으로

dt.datetime.strptime('2020-06-23','%Y-%m-%d')
```

```python
# 날짜 차이 구하기
# +,- *, / 다 된다
print(dt.datetime.today()-dt.timedelta(days=10)) # 오늘 날짜로부터 10일전 날짜 구하기
print(dt.datetime.strptime('2020-05-23','%Y-%m-%d')-dt.timedelta(days=10)) # 2020-05-23로부터 10일전 날짜 구하기 
```

```python
# 연습: 현재시간으로부터 2021년 크리스마스까지 남은 시간 출력

goal_time=dt.datetime(2021,12,25)
print('목표날짜: ',goal_time)
how_time=goal_time-dt.datetime.now()
print('날짜: ',how_time.days,'시간: ', how_time.seconds)

print('크리스마스까지 남은 날은 {}일 {}시간입니다.'.format(how_time.days,how_time.seconds//3600))
```

# pandas에서의 datetime

```python
import pandas as pd
```

```python
# 예제 파일 생성

data_t2=pd.date_range('2019-12-31',periods=12,freq='MS')
data_t3=pd.date_range('2019-01-01',periods=12,freq='M')
data_t=pd.concat([pd.DataFrame(data_t2),pd.DataFrame(data_t3)])
data_t.columns=['년도']
data_t
```

```python
# 년/월/일만 추출

t1=data_t['년도'].dt.year # 년도만 추출
t2=data_t['년도'].dt.month # 월만 추출
t3=data_t['년도'].dt.day # 일만 추출
print(t1.drop_duplicates()) # 중복제거
print(t2.drop_duplicates())
print(t3.drop_duplicates())
```

```python
data_t['년']=t1
data_t['월']=t2
data_t['일']=t3
data_t
```

```python
# 월, 일별 개수 확인

print(pd.DataFrame(data_t).groupby('월').count())
print(pd.DataFrame(data_t).groupby('일').count())
```

```python
# 분기 별로 분류

data_t['분기']=pd.to_datetime(data_t['년도'].values, format='%Y=%m').astype('period[Q]')
data_t
```

```python
# 날짜-시간 잘라내기
# 위의 데이터로 가정한다면 년도 컬럼이 인덱스가 되어야지 검색이 가능하다 
# brfore=a에서 a는 인덱스이다
# 또한 인덱스가 정렬이 되어야 한다

print(data_t2.truncate(before='2020-04')) # brfore=a : a 이전 제거
print(data_t2.truncate(after='2019-11')) # after=a : a 이후 제거
```

```python
# data_t의 년도 컬럼을 인덱스로 설정

data_t=data_t.set_index('년도')
data_t
```

```python
# 인덱스를 기준으로 정렬

data_t2=data_t.sort_index()
data_t2
```

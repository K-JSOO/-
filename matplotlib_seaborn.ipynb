# 파이썬 시각화
# 1. matplotlib

```python
# 패키지 import
# %matplotlib inline : 화면에 바로 그래프가 나오게 하는 코드\

import matplotlib.pyplot as plt
%matplotlib inline  
import numpy as np
import csv
import pandas as pd
import collections
```

```python
# 데이터 생성

x1=[i for i in range(1,11)]
y1=[2,3,4,6,7,9,10,16,17,20]

print(x1)
print(y1)
```

```python
# 라인 그래프

plt.plot(x1)
plt.plot(y1)
plt.plot(x1,y1)
plt.xlim(0,11) # x축 범위 설정
plt.ylim(0,20) # y축 범위 설정
plt.xlabel('x1') # x축 제목
plt.ylabel('y1') # y축 제목
plt.title('x1 and y1') # 그래프 제목
plt.legend(['x1','y1','x+y1']) # 여러개의 그래프를 표현 할 시 범례 표시
plt.show()
```

# ※ 색상, 마커, 선종류
1. 색상
    - blue :  	b
    - green :	g
    - red :	    r
    - cyan :    c
    - magenta :	m
    - yellow :	y
    - black :	k
    - white :	w                                                         
    

2. 마커
    - . :	point marker
    - , :	pixel marker
    - o :	circle marker
    - v :	triangle_down marker
    - ^ :	triangle_up marker
    - '>' :	triangle_right marker   # 싱글쿼테이션 제외
    - < :	triangle_left marker
    - 1 :	tri_down marker
    - 2 :	tri_up marker
    - 3 :	tri_left marker
    - 4 :	tri_right marker
    - s :	square marker
    - p :	pentagon marker
    - * :	star marker
    - h :	hexagon1 marker
    - H :	hexagon2 marker
    - + :	plus marker
    - x :	x marker
    - D :	diamond marker
    - d :	thin_diamond marker                                           
    

3. 선종류
    - '-' :	solid line style  # 싱글쿼테이션 제외
    - -- :	dashed line style
    - -. :	dash-dot line style
    - : :	dotted line style                                             
    

4. 스타일 설정
    - color(c) : 선 색깔
    - linewidth(lw)	: 선 굵기
    - linestyle(ls) : 선 스타일
    - marker : 마커 종류
    - markersize(ms) : 마커 크기
    - markeredgecolor(mec) : 마커 선 색깔
    - markeredgewidth(mew) : 마커 선 굵기
    - markerfacecolor(mfc) : 마커 내부 색깔

```python
# 그래프 색상과 마커

plt.plot(x1,'cs-') # 청록색 + 사각형 마커+ 실선
plt.plot(y1,'m.--') # 마젠타색 + 점 마커 + --선
plt.plot(x1,y1,'k+:') # 검은색 + 플러스 마커 + 도트선
plt.xlim(0,11)
plt.ylim(0,20)
plt.xlabel('x1')
plt.ylabel('y1')
plt.title('x1 and y1')
plt.legend(['x1','y1','x+y1'])
plt.show()
```

```python
# 막대그래프

plt.bar(x1,y1,color='skyblue')
plt.xlabel('x1')
plt.ylabel('y1')
plt.title('x1 and y1')
plt.show()
```

```python
# pie 그래프

a=[1,2,3,1,3,3,3,1,3]
a2=collections.Counter(a)  # 빈도수 세는 함수
print('a2 : ',a2)
a3=[i for i in a2.values()]
print('a3 : ',a3)
plt.pie(a3,labels=a2.keys(),colors=['pink','skyblue','lightgreen'])
plt.title('pie plot')
```

```python
# 히스토그램

np.random.seed(0)
x2=np.random.randn(1000) # 랜덤으로 1000개 추출
plt.grid() # 배경 눈금
plt.hist(x2,bins=20,alpha=0.6,rwidth = 0.9)
# bin: 구간개수, alpha : 투명도 (0:투명,1:불투명), rwidth: 그래프두께
```

```python
# 여러 그래프를 한 화면에 출력 -> subplots 사용

def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

def g(t):
    return np.sin(np.pi*t)

t1 = np.arange(0.0, 5.0, 0.01)
t2 = np.arange(0.0, 5.0, 0.01)

plt.subplot(221)  # 2x2 행렬의 1번째
plt.plot(t1, f(t1))

plt.subplot(222)  # 2x2 행렬의 2번째
plt.plot(t2, g(t2))

plt.subplot(223)  # 2x2 행렬의 3번째
plt.plot(t1, f(t1), 'r-')

plt.subplot(224)  # 2x2 행렬의 4번째
plt.plot(t2, g(t2), 'r-')

plt.show()
```

# 2. seaborn
- matplotlib를 기본으로 통계분석 목적으로 그래프를 더 쉽게 그리도록 래핑한 모듈

```python
import seaborn as sns
```

```python
ins=pd.read_csv('C:\\data\\insurance.csv')
ins
```

```python
# 산포도 그래프

sns.relplot(x='age',y='expenses',hue='sex',data=ins)
```

```python
# categorical scatterplot
# 분석하고자 하는 데이터가 카테고리식으로 분류가 가능하다면 함수 사용

sns.catplot(x='region',y='expenses',hue='sex',data=ins)
```

```python
# 라인 그래프

sns.lineplot(x='age',y='expenses',hue='sex',data=ins) # hue : 카테고리 값에 의한 실수값이 변화 측정 가능
```

```python
# 막대그래프

sns.barplot(x='region',y='children',data=ins)
```

```python
# 세부 집단 확인

sns.barplot(x='region',y='children',data=ins,hue='smoker')
```

```python
# 히스토그램 
# matplotlib의 hist보다 많이 쓰이는 시각화 방법

sns.distplot(x2,bins=20,kde=True) # 막대그래프도 무조건 같이 나온다
'''
ked_kws={}를 해주면 밀도 설정이 가능하다
ex) sns.distplot(x2,bins=20,kde=True,ked_kws={'color':'g','alpha':0.5,'linewidth':2})
'''
```

```python
# distplot 설정

sns.distplot(x2,bins=20,hist=True,kde=False,rug=False)
# kde=False : 막대그래프만 표시하겠다
```

```python
sns.distplot(x2,bins=20,hist=True,kde=False,rug=True)
# rug : x축에 데이터를 1차원으로 표시해줌
```

```python
# 카운드 플롯
# 각 카테고리 값별로 데이터가 얼마나 있는지 표시가능

sns.countplot(x='smoker',data=ins,palette=["C1", "k"])
```

```python
# joint plot
# 두 데이터가 연속형 변수라면 상단부분과 오른쪽 부분에 히스토그램이 같이 출력

sns.jointplot(x='age',y='expenses',data=ins)
```
